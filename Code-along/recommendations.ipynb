```python
nup_books = books.drop_duplicates("Book-Title")
nup_books.info()
```

```python
ratings_with_name = ratings.merge(nup_books, on="ISBN")
ratings_with_name.drop(["ISBN", "Image-URL-S", "Image-URL-M", "Image-URL-L", "Publisher"], axis=1, inplace=True)
ratings_with_name.head()
```

```python
len(ratings_with_name["User-ID"].unique())
```

```python
ratings_with_name.info()
```

```python
ratings_with_name.dropna(inplace=True)
ratings_with_name.info()
```

```python
x = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 100
expert_users = x[x].index
```

```python
filtered_user_ratings = ratings_with_name[ratings_with_name["User-ID"].isin(expert_users)]

y = filtered_user_ratings.groupby("Book-Title").count()["Book-Rating"] > 50
famous_books = y[y].index

famous_ratings = filtered_user_ratings[filtered_user_ratings["Book-Title"].isin(famous_books)]
```

```python
design_matrix = famous_ratings.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
```

```python
design_matrix.head()
```

```python
design_matrix.fillna(0, inplace=True)
```

```python
design_matrix.describe()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

scaler = StandardScaler()
design_matrix_scaled = scaler.fit_transform(design_matrix)
design_matrix_scaled
```

```python
similarity_score = cosine_similarity(design_matrix_scaled)
similarity_score
```

```python
import numpy as np

def recommend(book_name):
    index = np.where(design_matrix.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:6]
    data = []

    for index, similarity in similar_books:
        item = []
        temp_df = nup_books[nup_books["Book-Title"]==design_matrix.index[index]]
        item.extend(temp_df["Book-Title"].values)
        item.extend(temp_df["Book-Author"].values)
        data.append(item)
    return data
```

```python
recommend("The Hot Zone")
```

```python
from sklearn.neighbors import KNeighborsTransformer
k= 5

model = KNeighborsTransformer(n_neighbors=k, mode="distance", metric="cosine")

transform = model.fit_transform(design_matrix_scaled)

book_title= "The Hot Zone"
book_index = np.where(design_matrix.index==book_title)[0][0]
csr = transform[book_index]

print(csr)
```
